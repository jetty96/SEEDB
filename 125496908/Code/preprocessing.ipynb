```python
import psycopg2
import pandas as pd
```

```python
headers = ["age", "workclass", "fnlwgt", "education", "education_num", "marital_status",
           "occupation", "relationship", "race", "sex", "capital_gain", "capital_loss integer", 
           "hours_per_week", "native_country", "salary_range"]
adult_data_df = pd.read_table("./Data/adult.data.txt",  names = headers, delimiter=",")
```

```python
adult_data_df
```

```python
adult_data_df = adult_data_df.drop(['fnlwgt'], axis=1)
adult_data_df
```

```python
#Preprocessing
avg_age = adult_data_df["age"].mean()
adult_data_df["age"] = adult_data_df["age"].replace("?", avg_age)
```

```python
#mode = adult_data_df["workclass"].mode()
mode = adult_data_df.mode()["workclass"][0]
mode
adult_data_df["workclass"] = adult_data_df["workclass"].replace([" ?"], mode)
#adult_data_df["workclass"].head(30)
```

```python
#mode = adult_data_df["workclass"].mode()
mode = adult_data_df.mode()["education"][0]
mode
adult_data_df["education"] = adult_data_df["education"].replace([" ?"], mode)
#adult_data_df["workclass"].head(30)
```

```python
#mode = adult_data_df["workclass"].mode()
mode = adult_data_df.mode()["occupation"][0]
mode
adult_data_df["occupation"] = adult_data_df["occupation"].replace([" ?"], mode)
#adult_data_df["workclass"].head(30)
```

```python
#mode = adult_data_df["workclass"].mode()
mode = adult_data_df.mode()["race"][0]
mode
adult_data_df["race"] = adult_data_df["race"].replace([" ?"], mode)
#adult_data_df["workclass"].head(30)
```

```python
#mode = adult_data_df["workclass"].mode()
mode = adult_data_df.mode()["sex"][0]
mode
adult_data_df["sex"] = adult_data_df["sex"].replace([" ?"], mode)
#adult_data_df["workclass"].head(30)
```

```python
#mode = adult_data_df["workclass"].mode()
mode = adult_data_df.mode()["native_country"][0]
mode
adult_data_df["native_country"] = adult_data_df["native_country"].replace([" ?"], mode)
#adult_data_df["workclass"].head(30)
```

```python
avg_cap_gain = adult_data_df["capital_gain"].mean()
adult_data_df["capital_gain"] = adult_data_df["capital_gain"].replace([" ?"], avg_cap_gain)
```

```python
avg_cap_loss_int = adult_data_df["capital_loss integer"].mean()
adult_data_df["capital_loss integer"] = adult_data_df["capital_loss integer"].replace([" ?"], avg_cap_loss_int)
```
